```python
import os
import pandas as pd

# 현재 디렉토리에서 파일 목록 가져오기
try:
    files = os.listdir()
    # Excel 파일만 필터링
    excel_files = [file for file in files if file.endswith('.xlsx')]

    # 파일별 시트명 가져오기
    sheets_dict = {}
    for file in excel_files:
        try:
            # Excel 파일에서 시트명 가져오기
            excel_file = pd.ExcelFile(file)
            sheets_dict[file] = excel_file.sheet_names
        except Exception as e:
            sheets_dict[file] = f"오류 발생: {e}"

    # 결과 출력
    for file, sheets in sheets_dict.items():
        print(f"파일: {file}, 시트명: {sheets}")

except Exception as e:
    print("오류가 발생했습니다:", e)
```

```python
# 제공된 파일 및 시트 구조
file_sheets = {
    "김도연1.xlsx": ['op_earn', 'depre_amort', 'depre', 'age', 'n_acq_intangi'],
    "김도연2.xlsx": ['sales', 'tangi', 'intangi', 'cash'],
    "한민지.xlsx": ['amort', 'payab', 'invent', 'const_receiv2', 'n_acq_tangi'],
    "오영식.xlsx": ['receiv', 'st_debt', 'lt_debt', 't_debt', 'const_receiv1']
}

# 결과를 저장할 딕셔너리
dfs = {}

for file, sheets in file_sheets.items():
    for sheet in sheets:
        try:
            # 시트명을 키로 사용하여 데이터프레임 저장
            if sheet not in dfs:
                dfs[sheet] = pd.read_excel(file, sheet_name=sheet)
        except Exception as e:
            print(f"오류 발생 - 파일: {file}, 시트: {sheet}, 원인: {e}")

# 결과 확인
print("데이터프레임 딕셔너리가 생성되었습니다.")
for key, df in dfs.items():
    print(f"키: {key}, 데이터프레임 크기: {df.shape}")
```

```python
rows_to_drop = list(range(0, 13)) # 0 ~ 12

# 모든 코딩에서는 순서의 나열의 첫번째가 1이 아니라 0

for k in dfs.keys():
  dfs[k].columns = dfs[k].iloc[7] # firmcode line을 column명으로 지정
  dfs[k] = dfs[k].drop(rows_to_drop, axis=0)
  dfs[k] = dfs[k].set_index("Symbol")
```
