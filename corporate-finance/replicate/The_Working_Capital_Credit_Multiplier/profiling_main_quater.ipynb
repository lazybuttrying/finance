```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-darkgrid')

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['axes.titlepad'] = 20

# pd.set_option('display.float_format', lambda x: '%.6f' % x)

df = pd.read_csv('result_5y.csv',
    dtype={'ind_code': str, 'main_quater':int, 'cashflow_num': int
    }).drop(columns=[
        "quaters_num_cnt", "quaters_num_str", "quater_num", "quater_all", "cashflow_sum", "cashflow_num"])
df = df[~df["ind_code"].isna()]
```

# Without NaN industry code

```python
df = df[~df["ind_code"].isna()]
```

```python
df.describe()
```

```python
df["ind_code"].value_counts(dropna=False).to_frame().sort_index().plot(
    kind='bar', figsize=(10,5), title='산업별 Frequency',
    xlabel='산업코드', ylabel='Frequency'
)
```

```python
df["exchange"].value_counts(dropna=False).to_frame()
```

```python
df[['ind_code', 'exchange']].groupby(['ind_code', 'exchange', ]).size().to_frame().plot(
    kind='bar', figsize=(15,5), title='산업별 거래소별 Frequency',
    xlabel='산업코드, 거래소', ylabel='Frequency', legend=False,
    rot=80, grid=True
)
```

```python
df[['exchange', 'main_quater']].groupby(['exchange', 'main_quater', ]).size().to_frame().plot(
    kind='bar', figsize=(15,4), title='거래소별 Main Quater별 Frequency',
    xlabel='거래소, Main Quater', ylabel='Frequency',
    rot=80
)
```

```python
df[['ind_code', 'main_quater']].groupby(['ind_code', 'main_quater', ]).size().to_frame().plot(
    kind='bar', figsize=(20,5), title='산업별 Main Quater별 Frequency',
    xlabel='산업코드, Main Quater', ylabel='Frequency',
    rot=80
)
```
